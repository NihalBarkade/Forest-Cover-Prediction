```python
def get_df_info(df):
    print("\n\033[1mShape of DataFrame:\033[0m ", df.shape)
    print("\n\033[1mColumns in DataFrame:\033[0m ", df.columns.to_list())
    print("\n\033[1mData types of columns:\033[0m\n", df.dtypes)
    
    print("\n\033[1mInformation about DataFrame:\033[0m")
    df.info()
    
    print("\n\033[1mNumber of unique values in each column:\033[0m")
    for col in df.columns:
        print(f"\033[1m{col}\033[0m: {df[col].nunique()}")
        
    print("\n\033[1mNumber of null values in each column:\033[0m\n", df.isnull().sum())
    
    print("\n\033[1mNumber of duplicate rows:\033[0m ", df.duplicated().sum())
    
    print("\n\033[1mDescriptive statistics of DataFrame:\033[0m\n", df.describe().transpose())

# Call the function
get_df_info(df)
```

```python
#Splitting X and y
X = df.drop('Cover_Type',axis=1)
y = df['Cover_Type']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# Train-Test Split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
lg = LogisticRegression()
lg.fit(X_train,y_train)
y_pred_lg = lg.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_lg)
```

```python
print(classification_report(y_test,y_pred_lg))
```

```python
confusion_matrix(y_test,y_pred_lg)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred_dt = dt.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_dt)
```

```python
print(classification_report(y_test,y_pred_dt))
```

```python
confusion_matrix(y_test,y_pred_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred_rfc = rfc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_rfc)
```

```python
print(classification_report(y_test,y_pred_rfc))
```

```python
confusion_matrix(y_test,y_pred_rfc)
```

```python
from sklearn.ensemble import ExtraTreesClassifier


etc = ExtraTreesClassifier()
etc.fit(X_train, y_train)
y_pred_etc = etc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_etc)
```

```python
print(classification_report(y_test,y_pred_etc))
```

```python
confusion_matrix(y_test,y_pred_etc)
```

```python
#from sklearn.externals import joblib  # Use `import joblib` if using newer versions
import joblib  # Use this if the above import gives an error

# Dictionary to store models and their accuracies
models = {
    "Logistic Regression": (lg, accuracy_score(y_test, y_pred_lg)),
    "Decision Tree": (dt, accuracy_score(y_test, y_pred_dt)),
    "Random Forest": (rfc, accuracy_score(y_test, y_pred_rfc)),
    "Extra Trees": (etc, accuracy_score(y_test, y_pred_etc))
}

# Get the best model
best_model_name, (best_model, best_accuracy) = max(models.items(), key=lambda x: x[1][1])

print(f"\nBest Model: {best_model_name} with Accuracy: {best_accuracy:.2%}")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
print("Best model saved as 'best_model.pkl'")
```

```python
loaded_model = joblib.load("best_model.pkl")

# Predict using the loaded model
y_pred_loaded = loaded_model.predict(X_test)
print("Loaded model accuracy:", accuracy_score(y_test, y_pred_loaded))
```
